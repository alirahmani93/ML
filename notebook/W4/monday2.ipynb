```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../../../../../Downloads/Datasets/archive/CAvideos.csv')
```

```python
likes_=df.groupby(['category_id']).likes.agg('mean')
dislikes_ = df.groupby(['category_id']).dislikes.agg('mean')
views_=df.groupby(['category_id']).views.mean()

like_per_dis = likes_ / dislikes_
```

```python
like_per_view = likes_ / views_
```

```python
x=like_per_dis * like_per_view
x.plot(kind='bar')
```
